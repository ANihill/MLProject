## Clean DataSet

To clean the dataset we first need to identify irregular data.

```python
# https://towardsdatascience.com/data-cleaning-in-python-the-ultimate-guide-2020-c63b88bf0a0d
df['power'].hist(bins=100)
plt.show()
```

The dataset contains a number of entries where the power output is equal to 0. This occurs at various intervals right throughout the dataset, as seen below. This indicates that the power output may not have been recorded at these speeds. The first step will be to remove these erroneous entries.

```python
df.loc[df['power'] == 0]
```

```python
df = df.loc[df['power'] != 0]
df
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

## Training

```python
# https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
X = df['speed'].values
y = df['power'].values

X = X.reshape(-1,1)

# Split dataset into training set and test Set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state=0)
```

## Regression


```python
# https://github.com/ianmcloughlin/jupyter-teaching-notebooks/blob/master/regression.ipynb
m = lm.LinearRegression()
```

```python
m.fit(X_train, y_train)
```

```python
y_pred = m.predict(X_test)
```

```python
# https://scikit-learn.org/stable/auto_examples/linear_model/plot_ols.html
# The coefficients
print('Coefficients: \n', m.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```

## Keras

```python
#https://www.tensorflow.org/tutorials/keras/regression
train_dataset = df.sample(frac=0.7, random_state=0)
test_dataset = df.drop(train_dataset.index)
```

```python
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('power')
test_labels = test_features.pop('power')
```

```python
# Normalisation
train_dataset.describe().transpose()[['mean', 'std']]
```

```python
normalizer = preprocessing.Normalization()
```

```python
normalizer.adapt(np.array(train_features))
print(normalizer.mean.numpy())
```

```python
speed = np.array(train_features['speed'])

speed_normalizer = preprocessing.Normalization(input_shape=[1,])
speed_normalizer.adapt(speed)
```

```python
def build_and_compile_model(norm):
  model = kr.Sequential([
      norm,
      layers.Dense(64, activation='sigmoid'),
      layers.Dense(64, activation='sigmoid'),
      layers.Dense(64, activation='sigmoid'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001), metrics=[tf.keras.metrics.Accuracy()])
  return model
```

```python
model = build_and_compile_model(speed_normalizer)
```

```python
model.summary()
```

```python
#model.compile(
#    optimizer=tf.optimizers.Adam(learning_rate=0.1),
#    loss='mean_absolute_error', metrics='accuracy')
```

```python
%%time
history = model.fit(
    train_features['speed'], train_labels,
    epochs=500,
    # Calculate validation results on 20% of the training data
    validation_split = 0.2)
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [POWER]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
test_results = {}

test_results['model'] = model.evaluate(
    test_features['speed'],
    test_labels, verbose=0)
```

```python
x = tf.linspace(0.0, 25, 50)
y = model.predict(x)
```

```python
def plot_speed(x, y):
  plt.scatter(train_features['speed'], train_labels,  c='green',  label='Data')
  plt.plot(x, y, 'rx', label='Predictions')
  plt.xlabel('Speed')
  plt.ylabel('Power')
  plt.legend()

plot_speed(x,y)
```
